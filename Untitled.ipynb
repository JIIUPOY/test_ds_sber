```python
conn = sqlite3.connect(':memory:')
pays_dt.to_sql('table1', conn, index=False, if_exists='replace')
inn_info_public_dt.to_sql('table2', conn, index=False, if_exists='replace')

query = """
SELECT
    t1.hash_inn_kt hash_inn_kt,
    t1.hash_inn_dt hash_inn_dt,
    t1.week week,
    t1.count count,
    t1.sum sum,
    t2.hash_inn hash_inn_for_kt,
    t2.okved2 okved2_kt,
    t2.region region_kt,
    t2.is_public is_public_kt,
    t3.hash_inn hash_inn_for_dt,
    t3.okved2 okved2_dt,
    t3.region region_dt,
    t3.is_public is_public_dt
FROM table1 t1
INNER JOIN table2 t2
ON t1.hash_inn_kt = t2.hash_inn
INNER JOIN table2 t3
ON t1.hash_inn_dt = t3.hash_inn
"""
result = pd.read_sql(query, conn)
```
